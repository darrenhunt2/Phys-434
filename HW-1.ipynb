```python
# imports as necessary
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

# this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (15,10)
```

```python
# generate 100000 points for a normal distribution using scipy
points = stats.norm.rvs(loc = 5., scale = 0.01, size = 100000)
```

```python
# plot a histogram of the points using matplotlib
fig, ax = plt.subplots(1, 1)  # one subplot (coordinates 1,1) -- change subplots for more plots
ax.hist(points,50, density=True)  # plot a histogram using the 100,000 points we generated above
# 50 bins, or classes, where the points are put into. these are equal widths
# density=True returns a probability density
plt.tick_params(labelsize = 24)  # changes the font size of the label
plt.xlim([4.95,5.05])  # changes the limits of the x axis on either side that display on the subplot
x = np.linspace(4.95,5.05,1000)  # returns 1000 equally space points between 4.95 and 5.05
ax.plot(x,stats.norm.pdf(x,loc = 5., scale = 0.01),linewidth = 8,alpha = 0.7)  # draws a normally
# distributed line over the same subplot using equally spaced points
plt.show()  # makes plot output visible
```

```python
ax = plt.hist(points,50)  # using our initially generated set of points, plot a histogram w/ 50 bins
plt.yscale('log')  # the y axis is log scale -- semilog plot.
plt.tick_params(labelsize = 24)  # change font size of labels
plt.xlim([4.95,5.05])  # change x axes limits
plt.show()  # show plot below this cell

# this is the same data as the previous plot (same set of points), but now
# plotted on a semilog, which may reveal a different shape for the distribution.
```

```python
fig, ax = plt.subplots(1, 1)  # same as first example
ax.hist(points,50, density=True)  # ditto
plt.yscale('log')  # the scale of y axis is logarithmic
plt.tick_params(labelsize = 24)  # same as first ex
plt.xlim([4.95,5.05])  # same as first ex
x = np.linspace(4.95,5.05,1000)  # generate another set of equally spaced points, 1000 btwn these lims
ax.plot(x,stats.norm.pdf(x,loc = 5., scale = 0.01),linewidth = 8,alpha = 0.7)  # now, plot a normally
# distributed line, over our other data. is the original data set actually Gaussian?
plt.show()  # same as previous
```
